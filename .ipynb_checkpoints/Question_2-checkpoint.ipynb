# Question 2: Is there a significant difference between local and international sales?
----------------------------------------------------------------

Null Hypothesis: $H_0: \mu_1 = \mu_2$

Alternative Hypothesis: $H_a: \mu_1 != \mu_2$

### Import packages

```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import custom_tools as ct
import MyConn as mc
import scipy.stats as scs
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd, MultiComparison
import statsmodels.api as sm
import seaborn as sns

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
## Connection to the database ##
myconn = mc.MyConn(filename='./Northwind_small.sqlite')
```

```python
## Show the tables within the sqlite database ##
myconn.database_tables()
```

```python
query = """
    SELECT o.ID as order_id, p.ProductName, od.Discount, od.UnitPrice, od.Quantity, c.CompanyName, c.City, c.Region, c.Country
    FROM [Order] as o
    JOIN Customer as c
    ON o.CustomerId = c.Id
    JOIN OrderDetail as od
    ON order_id =  od.orderId
    JOIN Product as p
    ON od.ProductId = p.Id;
"""
df = myconn.run_query(query, load_df=True)
myconn.close_connection()
df.shape
```

```python
df['total_price'] = (1 - df.Discount)*df.UnitPrice*df.Quantity
```

## ANOVA & Tukey tests on Quantities by Region

#### ANOVA

There is a difference in variance amongst the different quantities and the regions that buy products from the Northwind Company.  However, there does not seem to be a difference in quantities when we compare those ordered locally against all international orders.

```python
# ANOVA Test on regions with Quantity as the target
alpha = 0.05

formula = f'Quantity ~ C(Region) + C(City) + C(Country) + C(local)'
lm = ols(formula, df).fit()
table = sm.stats.anova_lm(lm)
table['reject_null'] = table['PR(>F)'] < alpha
pd.DataFrame.from_dict(table)
```

#### Tukey Comparison Test using Quantity as the target

```python
# Set up the data for comparison (creates a specialised object)
MultiComp = MultiComparison(df.Quantity,df.Region)
result = MultiComp.tukeyhsd()

# Show all pair-wise comparisons:
# Print the comparisons
df_result = pd.DataFrame(data=result._results_table.data[1:], columns=result._results_table.data[0])
print(df_result)
result.plot_simultaneous(comparison_name='North America', figsize=(13,8))
```

View the statical difference of each region compared to North America

```python
df_result[((df_result.group1=='North America')|(df_result.group2=='North America'))]
```

Our Tukey test suggests that of all the regions compared to North America, the significant statistical difference occurs when comparing Southern Europe to North America.  Besides this region, there does not seem to be a statistical difference based on the results of the Tukey Test.

## ANOVA and Tukey test on total_price as the target

#### ANOVA

There is a difference in variance amongst the different quantities and the regions that buy products from the Northwind Company

```python
# ANOVA Test on region
alpha = 0.05

formula = f'total_price ~ C(Region) + C(City) + C(Country) + C(local)'
lm = ols(formula, df).fit()
table = sm.stats.anova_lm(lm)
table['reject_null'] = table['PR(>F)'] < alpha
pd.DataFrame(table)
```

#### Tukey Comparison Test by Regions

```python
# Set up the data for comparison (creates a specialised object)
MultiComp = MultiComparison(df.total_price,df.Region)
result = MultiComp.tukeyhsd()
# Show all pair-wise comparisons:
# Print the comparisons
df_result = pd.DataFrame(data=result._results_table.data[1:], 
                         columns=result._results_table.data[0])
df_result
```

Look only at regions compared to North America 

```python
df_result[((df_result.group1=='North America')|(df_result.group2=='North America'))]
```

The results of our Tukey test suggests that the difference in sales of North America vs the other sales regions only appears when we compare North America to Southern Europe.

```python
result.plot_simultaneous(comparison_name='North America', figsize=(13,8))
```

#### Tukey Test on total price of Local vs International sales as a whole

```python
# Set up the data for comparison (creates a specialised object)
MultiComp = MultiComparison(df.total_price,df.local)
result = MultiComp.tukeyhsd()
# Show all pair-wise comparisons:
# Print the comparisons
df_result = pd.DataFrame(data=result._results_table.data[1:], 
                         columns=result._results_table.data[0])
df_result
```

```python
result.plot_simultaneous(comparison_name='Local')
```

## Conclusion

Based on the results of our test, there is a significant difference in quantities of products ordered and sales generated in foreign regions compared to North America, when we look at them region by regions(North America, Scadindavia, Western Europe, etc.)  This difference is only seen when we compare North America to Southern Europe.

When we look at the difference between North America and the cumulative results of every other region, there does not seem to be a significant differece.  There does seem to be a slight one, but not one that is large enough to be noteworthy, based on our graphs.
